## Open CV

### 이미지에 숫자 더하기

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# opencv를 이용해서 이미지 데이터를 읽으면 numpy 의 ndarray 로 리턴
# 흑백은 2차원 컬러는 3차원
src1=cv2.imread('./data/lena.jpg',cv2.IMREAD_GRAYSCALE)
src2=np.zeros(shape=(512,512),dtype=np.uint8)+100
print(src1)

dst=src1+src2 # 뎃셈을 하면 오버플로우 가능성 존재
dst=cv2.add(src1,src2) # # OpenCV 함수로 덧셈을 하면 256 이상의 값은 255로 바꿔줌. 

# 이미지의 색상값은 0~255 까지여야 하는데 덧셈하면 256 이상의 값은 256 으로 나눈 나머지 값으로 가지게 된다. 
plt.imshow(dst,cmap='gray')
plt.axis('off')
plt.show()
```

### 수학 및 통계 함수

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# opencv를 이용해서 이미지 데이터를 읽으면 numpy 의 ndarray 로 리턴
# 흑백은 2차원 컬러는 3차원
src1=cv2.imread('./data/minMax.jpg',cv2.IMREAD_GRAYSCALE)

(min_val,max_val,_,_)=cv2.minMaxLoc(src1) # 최대 최소 값 할당

ratio=255/(max_val-min_val)
dst=(np.round(src1-min_val)*ratio).astype('uint8')

plt.imshow(src1,cmap='gray')
plt.imshow(dst,cmap='gray')
plt.axis('off')
plt.show()
```
