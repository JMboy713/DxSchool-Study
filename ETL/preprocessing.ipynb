# 데이터 전처리

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('./data 4/auto-mpg.csv',header=None)
df.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']
```

```python
print(df['horsepower'].unique()) # ? 라는 값이 존재해서 object 로 출력.

# ? 라는 값을 가진 데이터를 어떻게 할 것인지 결정.
# ? 값을 None 으로 설정하고 None 인 데이터를 제거
df['horsepower'].replace('?',np.nan,inplace=True)
df.dropna(inplace=True)
df
```

```python
# 이상치 제거후 자료형 변경

df['horsepower'] = df['horsepower'].astype('float')
df.info()
```

* origin 이라는 열은 1,2,3 값만 소유하고 있다. 
* 이 값은 실제로는 미국,유럽,일본을 의미하는 숫자이다.
* 분석을 할 때는 미국,유럽,일본으로 보여지는 것이 더 나을수도 있다.
* 이런 경우는 새로운 컬럼을 추가해서 보여지도록 할 수도 있고 컬럼의 값ㅇ르 변경할 수도 있다. 
* 범주형으로 변경해 두는 것이 분석 속도를 높이는데 도움이 된다.

```python
df['origin'].replace({1:'미국',2:'유럽',3:'일본'},inplace=True)
df['origin']=df['origin'].astype('category')
df.info() # 범주형으로 변경
# 이후 상관계수를 나타내면 category 형 변수는 빠지게 된다. 
```
